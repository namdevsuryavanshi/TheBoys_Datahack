```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page you want to scrape
url = "https://en.wikipedia.org/wiki/Film"


# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the HTML element that contains the main content of the page
content = soup.find(id='mw-content-text').get_text()

# Print the content as a string
string = (str(content))
```

```python


import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# Example string
text = string2

# Tokenize the string into words
words = text.split()

# Remove stop words from the words list
filtered_words = [word for word in words if word.lower() not in stop_words]

# Print the filtered words
type(filtered_words)

import pandas as pd

my_list = filtered_words


# Create a dictionary to store the frequency count of each element in the list
freq_dict = {}
for element in my_list:
    if element in freq_dict:
        freq_dict[element] += 1
    else:
        freq_dict[element] = 1

# Convert the dictionary to a Pandas DataFrame
df = pd.DataFrame.from_dict(freq_dict, orient='index', columns=['Frequency'])

# Sort the DataFrame in descending order by frequency
df = df.sort_values(by=['Frequency'], ascending=False)
df.head(30)
```

```python
import spacy

# Load the pre-trained English language model
nlp = spacy.load('en_core_web_sm')

# Define the string you want to analyze
text = "Python is a popular programming language."

# Analyze the text with spaCy
doc = nlp(text)

# Loop through the tokens in the document and print their meanings
for token in doc:
    print(token.text, token.pos_, token.lemma_, token.dep_)
```
